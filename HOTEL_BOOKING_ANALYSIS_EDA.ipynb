### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Creating a copy so as not to disturb original dataset
hotel_df_copy = hotel_df.copy()
````

**DATA CLEANING/MANIPULATIONS**

Cleaning data is crucial step before EDA as it will remove the ambigous data that can affect the outcome of EDA.

While cleaning data we will perform following steps: \
1. Drop duplicate rows \
2. Handling missing values. \
3. Convert columns to appropriate datatypes. \
4. Adding important columns

**Step 1: Drop duplicate rows**

````python
# droping duplicate values using drop_duplicates method
hotel_df_copy.drop_duplicates(inplace = True)
````

As there are 31994 duplicate rows in the dataset, so we are simply dropping all the duplicate rows from our dataset

**Step2: Handling missing values**

````python
# Columns having missing values.
hotel_df_copy.isnull().sum().sort_values(ascending = False)[:5] # checking top 5 columns having NaN values
````

Looks like we have quite alot of NaN values in company and agent. Highly doubt they have an impact on the analysis and will remove the 2 columns. So removing columns will be a better idea since those 2 attributes (agents and company) are unimportant.

````python
# droping 'agent' and 'company' columns from dataset using 'drop()' method
hotel_df_copy = hotel_df_copy.drop(columns = ['agent', 'company'])
````

````python
hotel_df_copy['children'].unique()
````

````python
# checking value count of 'children' column
hotel_df_copy['children'].value_counts()
````

This column 'children' has 0 as value which means 0 children were present in group of customers who made that transaction.
So, 'nan' values are the missing values due to error of recording data.

Since mode of the children column is '0.0', So we will replace all NaN values with '0.0'

````python
hotel_df_copy['children'].fillna(value = 0.0, inplace = True)
````

Next column with null value is 'country'. Since, this column has datatype of string. We will replace the null values with the 'other' as we do not have much null values compare size of the dataset.

````python
# replacing nan values with 'other' in country column
hotel_df_copy['country'].fillna(value = 'other', inplace = True)
````

````python
# Checking if all null values are removed
hotel_df_copy.isnull().sum().sort_values(ascending = False)[:6]
````

````python
# checking if we have any rows with total number of adults, children and babies equal to zero
hotel_df_copy[hotel_df_copy['adults']+hotel_df_copy['babies']+hotel_df_copy['children'] == 0].shape
````

we will remove such rows because it does not make any sense that the booking was done for no one.

````python
hotel_df_copy.drop(hotel_df_copy[hotel_df_copy['adults']+hotel_df_copy['babies']+hotel_df_copy['children'] == 0].index, inplace = True)
````

**Step 3: Converting columns to appropriate datatypes.**

Since 'children' and 'reservation_status_date' columns are not in correct datatype formate, we will converting them into appropriate datatype formates.

````python
# Converting datatype of column 'children' from float to int.
hotel_df_copy['children'] = hotel_df_copy['children'].astype('int64')

# changing datatype of column 'reservation_status_date' to data_type.
hotel_df_copy['reservation_status_date'] = pd.to_datetime(hotel_df_copy['reservation_status_date'], format = '%Y-%m-%d')
````

**Step 4: Adding important columns**

we will adding two coulumns which will be required to analyse and get some insights from the dataset,

first one is, total_people = adults + children + babies

second one, total_stay = stays_in_week_nights + stays_in_weekend_nights

````python
# adding column of total_people
hotel_df_copy['total_people'] = hotel_df_copy['adults']+hotel_df_copy['children']+hotel_df_copy['babies']

# adding colum of total stay
hotel_df_copy['total_stay'] = hotel_df_copy['stays_in_weekend_nights']+hotel_df_copy['stays_in_week_nights']

# creating function to check if same room type not alloted

def check_room_allot(x):
  if x['reserved_room_type'] != x['assigned_room_type']:
    return 1
  else:
    return 0

# applying lamda funtion to create one column of "same_room_not_alloted"
hotel_df_copy['same_room_not_alloted'] = hotel_df_copy.apply(lambda x : check_room_allot(x), axis = 1)
````

````python
# checking if columns are added
hotel_df_copy.columns
````

**Step 5: cleaning wrong filled data**

````python
hotel_df_copy['adr'].describe()
````

#### Chart - 15 Monthly Hotel Booking Trends

````python
# Chart - 15 visualization code
# Checking monthly hotel booking trends

# Creating list of months
month_order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
df['arrival_date_month'] = pd.Categorical(df['arrival_date_month'], categories=month_order, ordered=True)

# grouping the data by monthly booking
monthly_bookings = df.groupby('arrival_date_month').size()

# visualization using line plot
plt.figure(figsize=(14, 6))
sns.lineplot(x=monthly_bookings.index, y=monthly_bookings.values, marker='o', linestyle='-', color='teal', markersize=10, linewidth=2.5)
plt.title('Monthly Hotel Booking Trends', fontsize=18, fontweight='bold')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Bookings', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.gca().spines['top'].set_color('none')
plt.gca().spines['right'].set_color('none')
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A line plot was chosen to clearly show the trend of hotel bookings over the months, making it easy to identify peak and off-peak periods.

##### 2. What is/are the insight(s) found from the chart?

The chart shows a clear seasonal trend in hotel bookings. January has the lowest number of bookings, followed by November and December. There's a gradual increase in bookings from February, peaking in August, which is the highest booking month. May, June, and July also experience high booking volumes. After August, there's a noticeable decrease in bookings towards the end of the year.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

These insights are crucial for strategic planning and resource management. The peak season from May to August suggests the need for increased staffing and resources to accommodate the high number of guests. Promotional activities and special offers could be targeted in the low-booking months like January and November to boost occupancy. Understanding these trends can also aid in revenue management, allowing the hotel to optimize pricing during high-demand periods and offer discounts or packages during off-peak times to maintain a steady flow of guests.

#### Chart - 16 Monthly Trends in the Average Daily Rate (ADR)

````python
# Chart - 16 visualization code
# Checking Monthly Trends in the Average Daily Rate (ADR)

# Capping the 'adr' values at a maximum of 500 to remove extreme outliers
df['adr'] = df['adr'].clip(upper=500)

# This helps in displaying the months in a chronological order on the x-axis
reindex = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df['arrival_date_month'] = pd.Categorical(df['arrival_date_month'], categories=reindex, ordered=True)

# Visualization using boxplot
# Using Seaborn for a more aesthetically pleasing boxplot
plt.figure(figsize=(14, 6))
sns.boxplot(x='arrival_date_month', y='adr', data=df, palette="viridis")
plt.title('Monthly ADR Distribution', fontsize=16, fontweight='bold', color='navy')
plt.xlabel('Month', fontsize=14, color='darkgreen')
plt.ylabel('Average Daily Rate (ADR)', fontsize=14, color='darkgreen')
plt.xticks(rotation=45, fontsize=12, color='maroon')
plt.yticks(fontsize=12, color='maroon')
plt.grid(True, linestyle='--', alpha=0.7)
plt.gca().spines['top'].set_color('none')
plt.gca().spines['right'].set_color('none')
plt.show()
````

##### 1. Why did you pick the specific chart?

The boxplot was chosen because it effectively displays the distribution of Average Daily Rates (ADR) across different months. It highlights the median, quartiles, and potential outliers, providing a clear picture of how the ADR varies. The boxplot is particularly useful for identifying trends, seasonal effects on pricing, and spotting any anomalies or outliers in the data.

##### 2. What is/are the insight(s) found from the chart?

The chart distinctly indicates that January experiences the lowest average daily rates (ADR), signaling a potential off-peak period for the hotel. A gradual increase in ADR is observed from February, continuing through March, April till August, suggesting a rising demand leading into the spring months. The ADR reaches its peak in August, which aligns with the summer holiday season when hotels often see increased bookings. Following this peak, there's a noticeable decline from September as the year progresses towards its end, with November and December again showing lower ADRs. This pattern highlights the peak season during the summer months of June, July, August, and September, while the winter months, particularly January, November, and December, are characterized by the lowest ADRs.
